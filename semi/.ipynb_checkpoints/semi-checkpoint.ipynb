```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 그룹 구분 컬럼 추가
df_oxi_good['Group'] = 'Good'
df_oxi_bad['Group'] = 'Bad'

# 데이터 병합
df_compare = pd.concat([df_oxi_good, df_oxi_bad], ignore_index=True)

cols_to_plot = ['Temp_OXid', 'ppm', 'Pressure', 'Oxid_time', 'thickness', 'Lot_Num', 'Wafer_Num']

# subplot 설정
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
axes = axes.flatten()

# 각 변수별로 good/bad 비교 boxplot
for i, col in enumerate(cols_to_plot):
    sns.boxplot(data=df_compare, x='Group', y=col, palette='Set2', ax=axes[i])
    axes[i].set_title(f'{col} Distribution (Good vs Bad)')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    
# 빈칸 비우기
for j in range(len(cols_to_plot), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 시간 순서 정렬
df_oxi = df_oxi.sort_values('Datetime')

# 챔버별 통계 계산
chamber_stats = df_oxi.groupby('Ox_Chamber')['thickness'].agg(['mean', 'std']).reset_index()
chamber_stats['ucl'] = chamber_stats['mean'] + 3 * chamber_stats['std']
chamber_stats['lcl'] = chamber_stats['mean'] - 3 * chamber_stats['std']

# 챔버별로 FacetGrid 생성
g = sns.FacetGrid(df_oxi, col='Ox_Chamber', col_wrap=3, height=4, sharey=True)

# 챔버별 산점도
g.map_dataframe(sns.scatterplot, x='Datetime', y='thickness', hue='Ox_Chamber', palette='Set2', s=30)

# 챔버별 기준선 (평균, UCL, LCL)
for ax, chamber in zip(g.axes.flat, chamber_stats['Ox_Chamber']):
    mean = chamber_stats.loc[chamber_stats['Ox_Chamber'] == chamber, 'mean'].values[0]
    ucl = chamber_stats.loc[chamber_stats['Ox_Chamber'] == chamber, 'ucl'].values[0]
    lcl = chamber_stats.loc[chamber_stats['Ox_Chamber'] == chamber, 'lcl'].values[0]
    
    ax.axhline(mean, color='blue', linestyle='--', linewidth=1)
    ax.axhline(ucl, color='red', linestyle='--', linewidth=1)
    ax.axhline(lcl, color='red', linestyle='--', linewidth=1)
    ax.tick_params(axis='x', rotation=45)
    ax.set_title(f"Chamber: {chamber}")

# 그래프 꾸미기
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Thickness Control Chart by Chamber', fontsize=14)
plt.show()
```
