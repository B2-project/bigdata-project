```python
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf

reg_model = smf.ols(formula = "Price ~ Location + Year + Kilometers_Driven + Fuel_Type + Transmission + Owner_Type + Mileage + Engine + Power + Seats + New_Price", data=df_with_dummies)
reg_result = reg_model.fit()
print(reg_result.summary())
```

```python
from statsmodels.api import qqplot, add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor

df_train_x = df_train.drop("Price", axis=1)
df_train_x_const = add_constant(df_train_x)

df_vif = pd.DataFrame()
df_vif["variable"] = df_train_x_const.columns
df_vif['VIF'] = [variance_inflation_factor(df_train_x_const.values,i)\
                for i in range(df_train_x_const.shape[1])]

df_vif.sort_values('VIF', ascending = True).round(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 숫자형 칼럼만 선택
numeric_columns = df.select_dtypes(include=['number']).columns

# 박스 플롯을 그릴 개수를 설정 (5개씩)
num_plots = 5
num_columns = len(numeric_columns)
num_rows = (num_columns // num_plots) + (1 if num_columns % num_plots != 0 else 0)

# 서브플롯 생성
fig, axes = plt.subplots(num_rows, num_plots, figsize=(15, num_rows * 6))
axes = axes.flatten()  # axes를 1D 배열로 변환

# 박스 플롯 그리기
for i, col in enumerate(numeric_columns):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f'Box plot of {col}')

# 레이아웃 자동 조정
plt.tight_layout()
plt.show()
```
